## Missing Data

In the Towards Data Science article, they list 3 techniques for detecting any missing data frame. Try all three before deciding on a way to solve any missing data issues.

```python
# Plot a heatmap of any missing data

cols = rat_sightings.columns[:] # first 30 columns
colors = ['#000099', '#ffff00'] # specify the colours - yellow is missing. blue is not missing.
sns.heatmap(rat_sightings[cols].isnull(), cmap=sns.color_palette(colors))
```

```python
# Try out a missing data percentage list! 

for col in rat_sightings.columns:
    pct_missing = np.mean(rat_sightings[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Plot a missing data histogram
for col in rat_sightings.columns:
    missing = rat_sightings[col].isnull()
    num_missing = np.sum(missing)
    
    if num_missing > 0:  
        print('created missing indicator for: {}'.format(col))
        rat_sightings['{}_ismissing'.format(col)] = missing


# then based on the indicator, plot the histogram of missing values
ismissing_cols = [col for col in rat_sightings.columns if 'ismissing' in col]
rat_sightings['num_missing'] = rat_sightings[ismissing_cols].sum(axis=1)

rat_sightings['num_missing'].value_counts().reset_index().sort_values(by='index').plot.bar(x='index', y='num_missing')
```

Now that you have tried all 3 techniques for detecting missing data, did you find any? Reading through the article, which solution do you think will help you clean the data set and handle the missing data? Try out your solution below!

```python
# Handle any missing data in your dataframe.

#Dropping Columns w/High Missing Data Percentage and Replacing Columns w/Low Missing Data Percentage :


for col in rat_sightings.columns:
    pct_missing = np.mean(rat_sightings[col].isnull())
    
    if pct_missing > .5:
        high_missing_data = rat_sightings.drop(col, axis=1)
        
        
#numeric         
df_numeric = rat_sightings.select_dtypes(include=[np.number])
numeric_cols = df_numeric.columns.values

for col in numeric_cols:
    missing = rat_sightings[col].isnull()
    num_missing = np.sum(missing)

    if num_missing > 0:  # only do the imputation for the columns that have missing values.
        print('imputing missing values for: {}'.format(col))
        rat_sightings['{}_ismissing'.format(col)] = missing
        med = rat_sightings[col].median()
        rat_sightings[col] = rat_sightings[col].fillna(med)        

# categorical

df_non_numeric = rat_sightings.select_dtypes(exclude=[np.number])
non_numeric_cols = df_non_numeric.columns.values

for col in non_numeric_cols:
    missing = rat_sightings[col].isnull()
    num_missing = np.sum(missing)

    if num_missing > 0:  # only do the imputation for the columns that have missing values.
        print('imputing missing values for: {}'.format(col))
        rat_sightings['{}_ismissing'.format(col)] = missing

        top = rat_sightings[col].describe()['top'] # impute with the most frequent value.
        rat_sightings[col] = rat_sightings[col].fillna(top)
```
